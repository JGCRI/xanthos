```python
import calendar
import os
import pkg_resources

import numpy as np
import pandas as pd
import xarray as xr

import xanthos

import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
def construct_days_per_month(start_yr, through_yr, yr_interval=1):
    """Construct an array of the number of days per month for all years when
    considering leap years.
    :param start_yr:                Four digit integer for year dataset starts in.
                                    Dataset must start on January 01.
    :param through_yr:              Four digit integer for year dataset covers through.
                                    Dataset must end covering December 31.
    :param yr_interval:             Integer for the year interval. Default: 1
    
    """
    
    DY_PER_MTH_FORMATS = {'standard': [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31],
                          'leap': [31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]}
    
    l = []
    for yr in range(start_yr, through_yr + yr_interval, yr_interval):

        if calendar.isleap(yr):
            days_per_month = DY_PER_MTH_FORMATS['leap']

        else:
            days_per_month = DY_PER_MTH_FORMATS['standard']

        l.extend(days_per_month)

    return np.array(l)
```

```python
model = 'gswp3'

output_dir = '/Users/d3y010/projects/xanthos/example/input/calibration'
output_file = os.path.join(output_dir, f'vic_{model}_nobc_hist_nosoc_co2_qtot_basin_km3-per-mth_1971_2010.csv')

# directory containing the runoff netcdf files from VIC
data_dir = f'/Users/d3y010/projects/xanthos/isimip/runoff/vic/{model}'

# runoff file list
file_list = [os.path.join(data_dir, i) for i in os.listdir(data_dir)]

# name of data variable
data_variable_name = 'qtot'

xanthos_grid_num = 67420

# xanthos reference file
ref_landcell_file = '/Users/d3y010/projects/xanthos/example/input/reference/xanthos_0p5deg_landcell_reference.csv'

# existing calibartion file
old_calib_file = '/Users/d3y010/projects/xanthos/example/input/calibration/vic_watch_basin_km3_1971_2001_monthly.csv'
odf = pd.read_csv(old_calib_file)
```

```python
# read in required reference data
ref_df = pd.read_csv(ref_landcell_file)

# generate landcell index array
lat_lon_array = ref_df[['latitude_index', 'longitude_index']].values

# get basin by landcell array
basin_id_array = ref_df[['basin_id']]['basin_id'].values

ref_df.set_index('grid_id', inplace=True)

grid_to_basin_dict = ref_df[['basin_id']].to_dict()['basin_id']

grid_to_hectare_dict = ref_df[['area_hectares']].to_dict()['area_hectares']

ref_df.reset_index(inplace=True)
```

```python
hold_dfs = []

for f in file_list:
    
    print(f"Processing:  {f}")
    
    basename = os.path.splitext(os.path.basename(f))[0]
    base_split = basename.split('_')
    start_yr = int(base_split[-2])
    end_yr = int(base_split[-1])
    n_months = (end_yr - start_yr + 1) * 12
    
    # n months per file
    months = list(range(1, n_months + 1, 1))

    days_per_month = construct_days_per_month(start_yr, end_yr)
    
    # read in NetCDF file
    ds = xr.open_dataset(f)

    # convert daily to monthly mean
    dsr = ds.resample(time='M').mean('time')

    data_arr = dsr.variables[data_variable_name][:]

    # set zeros array for output
    arr_out = np.zeros(shape=(xanthos_grid_num, data_arr.shape[0]))

    for index, coord_idx in enumerate(lat_lon_array):
        lat_idx, lon_idx = coord_idx

        # get Xanthos land cell value by corresponding index
        arr_out[index, :] = data_arr[:, lat_idx, lon_idx]

    # create dictionary of {month: data_array}
    d = {}
    for i in range(arr_out.shape[1]):
        v = arr_out[:, i]
        d[i+1] = v

    # build data frame 
    df = pd.DataFrame(d)

    df['grid_id'] = df.index + 1
    df['basin_id'] = df['grid_id'].map(grid_to_basin_dict)
    df['sqkm'] = df['grid_id'].map(grid_to_hectare_dict) * 0.01

    for i in months:

        # n days in the current year, month
        n_days = days_per_month[i-1]
        
        # n seconds per day * n days in month
        n_seconds_per_month = 86400 * n_days

        # kg m-2 s-1 to mm/month to km/month to km3/month
        df[i] = df[i] * n_seconds_per_month * 1e-6 * df['sqkm']


    df.drop(['grid_id', 'sqkm'], inplace=True, axis=1)

    # group by basin
    grp = df.groupby('basin_id').sum()

    grp.reset_index(inplace=True)
    
    dx = {'basin': [], 'year': [], 'month': [], 'q': []}

    for i in grp['basin_id'].values:

        yr = start_yr - 1

        for j in months:

            dx['basin'].append(i)

            mth = np.mod(j, 12)

            if mth == 0:
                mth = 12

            if mth == 1:
                yr += 1

            dx['year'].append(yr)

            dx['month'].append(mth)

            q = grp.loc[grp['basin_id'] == i][j].values[0]

            dx['q'].append(q)

    dfx = pd.DataFrame(dx)   
    
    hold_dfs.append(dfx)

fdf = pd.concat(hold_dfs)

fdf.sort_values(by=['basin', 'year', 'month'], inplace=True)

fdf.to_csv(output_file, index=False)
```

```python
# original from 2015 paper 
orig_watch = pd.read_csv(os.path.join(output_dir, 'vic_watch_basin_km3_1971_2001_monthly.csv'))

# new from isimip
wfdei = pd.read_csv(os.path.join(output_dir, 'vic_wfdei_nobc_hist_nosoc_co2_qtot_basin_km3-per-mth_1971_2010.csv'))
gswp3 = pd.read_csv(os.path.join(output_dir, 'vic_gswp3_nobc_hist_nosoc_co2_qtot_basin_km3-per-mth_1971_2010.csv'))
watch = pd.read_csv(os.path.join(output_dir, 'vic_watch_nobc_hist_nosoc_co2_qtot_basin_km3-per-mth_1971_2010.csv'))

```

```python
st_yr = 1970
en_yr = 2020

orig_watch_df = orig_watch.groupby(['year']).sum()
watch_df = watch.groupby(['year']).sum()
wfdei_df = wfdei.groupby(['year']).sum()
gswp3_df = gswp3.groupby(['year']).sum()

orig_watch_df.reset_index(inplace=True)
watch_df.reset_index(inplace=True)
wfdei_df.reset_index(inplace=True)
gswp3_df.reset_index(inplace=True)

orig_watch_df['dataset'] = 'orig_watch'
watch_df['dataset'] = 'watch'
wfdei_df['dataset'] = 'wfdei'
gswp3_df['dataset'] = 'gswp3'


mdf = pd.concat([orig_watch_df, watch_df, wfdei_df, gswp3_df])

mdx = mdf.pivot(index='year', columns='dataset', values='q')

dims = (11.7, 8.27)

fig, ax = plt.subplots(figsize=dims)

sns.lineplot(data=mdx)

ax.set_ylabel("km3/yr")
ax.set_title('Global Runoff')

```
